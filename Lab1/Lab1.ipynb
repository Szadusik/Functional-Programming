```haskell
vec3DLen :: (Double, Double, Double) -> Double
vec3DLen (x,y,z) = sqrt (x*x + y*y + z*z)
vec3DLen (1,1,1)
```

```haskell
swap :: (Int, Char) -> (Char, Int)
swap (x,y) = (y,x)
swap (1,'x')
```

```haskell
threeEqual :: (Int, Int, Int) -> Bool
threeEqual (x,y,z) = if (x/=y) || (y/=z)
                        then False
                        else True
                        
threeEqual(3,3,1)
```

```haskell
absInt :: Int -> Int
absInt x = if x > 0
            then x
            else (-1)*x
absInt (-3)
```

```haskell
min2Int :: (Int, Int) -> Int
min2Int (x,y) = if x > y
                    then y
                    else x
min2Int (-2,-1)                    
```

```haskell
min3Int :: (Int, Int, Int) -> Int
min3Int (x,y,z) 
    | (x<y) && (x<z) = x
    | (y<z) && (y<x) = y
    | otherwise = z
    
min3Int(14,16,3)
```

```haskell
sgn :: Int -> Int
sgn n
    | n>0 = 1         
    | n<0 = -1
    | otherwise = 0   
{-| Tutaj znajduje się wieloliniowy komentarz, Funkcja określająca wartość signum
Dla liczb większych od zera funkcja funkcja signum przyjmuje wartość równą 1
Dla liczb równych dokładnie 0 funkcja przyjmie wartość 0
W przeciwnym wypadku wartość funkcji będzie wynosiła -1
W taki sposób została opisana funkcja sgn -}
sgn 9
```

```haskell
or' :: (Bool, Bool) -> Bool
or' (True,True) = True
or' (True,False) = True
or' (False,True) = True -- Alternatywa
or' (False,False) = False
```

```haskell
and' :: (Bool, Bool) -> Bool
and' (True, True) = True
and' (False, True) = False -- Koniunkcja
and' (True, False) = False
and' (False, False) = False
```

```haskell
nand' :: (Bool, Bool) -> Bool
nand' (True, True) = False -- Negacja koniunkcji
nand' (True, False) = True
nand' (False, True) = True
nand' (False, False) = True
```

```haskell
xor' :: (Bool, Bool) -> Bool
xor' (True, True) = False
xor' (True, False) = True -- Alternatywa wykluczająca
xor' (False, True) = True
xor' (False, False) = False
```

```haskell
unitVec2D :: (Double, Double) -> (Double, Double)
unitVec2D (x,y) = (x/d,y/d)
                    where d = sqrt(x^2 + y^2)
unitVec2D (1,1)
```

```haskell
unitVec2D :: (Double, Double) -> (Double, Double)
unitVec2D (x,y) = let d = sqrt(x^2 + y^2)
                    in (x/d,y/d)
unitVec2D (1,1)
```

```haskell
roots :: (Double, Double, Double) -> (Double, Double)
roots (a, b, c) = ( (-b - d) / e, (-b + d) / e )
   where d = sqrt (b * b) - (4 * a * c)
         e = 2 * a
```
